#  remove duplicates and drop NaN values

```python
dict2={'name': ['salman', np.nan ,'farid'],
        'marks': [56,np.nan,89],
       'city': ['kol',np.nan,'mum']}
```

```python
dict2
```

```python
df=pd.DataFrame(dict2)
df
```

```python
df.dropna(how="all",axis=0)
```

```python
df.drop_duplicates(subset=['name'],keep='first')
```

```python
df['name'].value_counts(dropna=False)
```
